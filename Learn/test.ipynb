```python
import pandas as pd
import yfinance as yf
import datetime as dt
import numpy as np
```

```python
end = dt.datetime.now()
start = dt.datetime(2012,1,1)
start,end
```

```python
stockList = ['HDFCBANK','ICICIBANK','SBIN','KOTAKBANK']
stocks = [i + '.NS' for i in stockList]
stocks
```

```python
df = yf.download(tickers=stocks, start=start, end=end)
```

```python
df = df.stack(level='Ticker')
```

```python
df = df.unstack()
df.head()
```

```python
close = df.Close
close[close.index > end - dt.timedelta(days=100)].describe()
```

```python
import plotly.offline as pyo
pyo.init_notebook_mode(connected=True)

pd.options.plotting.backend = 'plotly'
```

```python
close.plot()
```

```python
close['SBIN.NS'].pct_change().plot(kind='hist',nbins=50)
```

```python
close['SBIN.NS'].pct_change().agg(['min','max'])
```

```python
test = close
test['shift'] = test['HDFCBANK.NS'].shift(1)
test.head()
```

```python
test.dropna(inplace=True)
```

```python
test['log_value'] = np.log(test['HDFCBANK.NS']/test['shift'])
test.head()
```

```python
test['change'] = test['log_value'].cumsum()
```

```python
test.head()
```

```python
test['change'].plot()
```

```python
test['change_pct'] = (np.exp(test['change'])-1)*100
test.head()
```

```python
test['change_pct'].plot()
```
