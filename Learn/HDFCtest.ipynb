```python
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.offline as plo
import datetime as dt
```

```python
plo.init_notebook_mode(connected=True)
pd.options.plotting.backend = 'plotly'
```

```python
endDate = dt.datetime.today()
startDate = dt.datetime(2012,1,1)
df = yf.download('HDFCBANK.NS',end=endDate,start=startDate)
```

```python
df.head()
```

```python

close = df.Close
```

```python
close.head()
```

```python
close.plot()
```

```python
close.columns = ['Close']
```

```python
close['Log_ret'] = np.log(close['Close']/close['Close'].shift(1))
```

```python
close.dropna(inplace=True)
close.head()
```

```python
close['Log_ret'].plot()
```

```python
close['Log_sum'] = close['Log_ret'].cumsum()
```

```python
close['Log_sum'].plot()
```

```python
closeMean = close['Close'].mean()
```

```python
close['Mean_sub'] = close['Close'] - closeMean
close['Sub_sq'] = np.square(close['Mean_sub'])
```

```python
subSum = close['Sub_sq'].sum()
volatility = (subSum/close.shape[0])**0.5
volatility
```

```python
close.head()
```

```python
close['Rolling_mean'] = close['Close'].rolling(14).mean()
```

```python
close['Rolling_sub'] = close['Close'] - close['Rolling_mean']
close['Rolling_sq'] = np.square(close['Rolling_sub'])
```

```python
close['Volatility'] = (close['Rolling_sq'].rolling(14).sum()/14)**0.5
```

```python
close['Volatility'].dropna().plot()
```
